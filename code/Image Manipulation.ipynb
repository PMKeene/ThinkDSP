```python
def roll_spectrum(spectrum):
    n, m = spectrum.shape
    spectrum = numpy.roll(spectrum, n//2, axis=0)
    spectrum = numpy.roll(spectrum, m//2, axis=1)
    return spectrum

def plot_spectrum(spectrum):
    plot_array(numpy.log(numpy.absolute(spectrum)))
```

```python
def HighPass(rolled, r):
    m,n=rolled.shape
    Filt=numpy.zeros_like(rolled)
    Filt[:,:]=rolled[:,:]
    for i in range(m-1):
        for j in range(n-1):
            if numpy.sqrt((i-m//2)**2 +(j-n//2)**2)<= (r):
                Filt[i,j]=0
#     Filt=roll_spectrum(rolled)
    return Filt  
```
