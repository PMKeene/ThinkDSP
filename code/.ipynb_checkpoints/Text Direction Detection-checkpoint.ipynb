# Text Direction Detection with Fourier Transforms:

There are many libraries that are slowly digitizing their collections of older or rarer books. Frequently this is done by using overhead book scanners that take a photo of each pair of pages as a person flips through, since splaying a book flat out for more traditional scanners can be quite harsh on the spine. Either way, however, sometimes some post-processing of these scans is desirable to correct for the off-kilter orientation of the book or the curvature of the page, especially as a precursor to further image processing to actually parse the text. 

This means we need a way to detect and correct the orientation of the text in these images--which I propose can be done with the Fourier Transform (FT) of the scan. Fortunately for us, printed text on a page tends to have a strong regularly repeating pattern in the direction along the spine of the book resulting from the rows of text, which will hopefully be somewhat less prominent in the perpendicular direction due to varying line and word lengths, as well as less uniform character width and shape. 

To test this we need an image of an open book, and a little boiler plate code to help us take and manipulate the FT.

```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import pandas
import scipy.signal

from IPython.display import Image, display
from matplotlib import pyplot
```

```python
pyplot.rcParams['image.cmap'] = 'gray'
pyplot.rcParams['image.interpolation']='nearest'
```

It would also be helpful to have some helper functions for plotting our images and spectrums, since we will want to center the symmetry in any spectrum we take and will find it most revealing to look at the spectrums in a log scale.

```python
def plot_array(array):
    pyplot.imshow(array)
    thinkplot.Show(legend=False)
    
def plot_arrays(array):
    thinkplot.PrePlot(cols=2)
    pyplot.imshow(array.real)
    thinkplot.SubPlot(2)
    pyplot.imshow(array.imag)
    thinkplot.Show(legend=False)
```

```python
import matplotlib.image
from skimage import img_as_float

def roll_spectrum(spectrum):
    n, m = spectrum.shape
    spectrum = numpy.roll(spectrum, n//2, axis=0)
    spectrum = numpy.roll(spectrum, m//2, axis=1)
    return spectrum

def plot_spectrum(spectrum):
    plot_array(numpy.log(numpy.absolute(spectrum)))
```

```python
Book1_orig = matplotlib.image.imread('Book1.jpeg')
Book1 = img_as_float(Book1_orig.sum(axis=2, dtype=numpy.float64))
```

```python
plot_array(Book1)
Book1.shape
```

Looking at our image, we might predict that in addition to the spacing of the lines, the paragraph breaks, the header, the edges of the book itself, and the closeness of the page to the camera at different parts of the photo will make for a number of different frequencies occuring in the long direction of the page. We would also expect there to be some power in the perpendicular direction, as well, especially from the edge of the book on the dark . 

In order to see if our predictions pan out, we need to take the two-dimensional FT.

```python
BookSpectrum = scipy.fftpack.fft2(Book1)
Centered = roll_spectrum(BookSpectrum)
```

```python
numpy.amax(Centered.real)
```

```python
plot_spectrum(Centered)
#plot_spectrum(Centered>.002*numpy.amax(Centered.real))
```

Immediately apparent to the wonderful pattern recognition software pre-installed in our skulls is a series of high values along a diagonal at the same angle as the book in the photo, with other high values along the perpendicular diagonal and the vertical and horizontal. However, actually filtering this spectrum to pick up on the diagonal we are interested in can be tricky, and involves fiddling with threshold values. 

The horizontal and vertical components, at least, are from the strong edge effects of the photo, where the FT implicitly assumed that the image was periodic when in fact it does not tile very smoothly. We can help get rid of these by cropping the image a little more closely on the actual page we are attempting to re-orient.

```python
SlicedBook=Book1[200:,100:]
plot_array(SlicedBook)
```

```python
SlicedBookSpectrum = scipy.fftpack.fft2(SlicedBook)
Centered2 = roll_spectrum(SlicedBookSpectrum)
plot_spectrum(Centered2)
```

Now the diagonal is even more apparent, and we are ready to try and pick it out from the rest of the spectrum by supressing the smaller values and taking a linear regression of the remaining points. 

```python
threshold_spectrum=numpy.copy(Centered2)
m,n=Centered2.shape
maximum=numpy.amax(Centered2)

for row in range(m-1):
    for column in range(n-1):
        if threshold_spectrum[row, column]<.002*maximum:
            threshold_spectrum[row,column]=0
            
plot_spectrum(threshold_spectrum)
x,y=numpy.nonzero(threshold_spectrum)
```

By tinkering with the threshold values, we are able isolate points that are primarily along the diagonal we wanted, but that is a major weakness in this process, as we cannot guarantee that we will want the same values for all lighting conditions, books, or even series of photographs from the same book.

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
print(slope)
angle=numpy.average(numpy.arctan2(y,x)*180/numpy.pi)#We could also just get the angle from the slope of the line of best fit
```

From the remaining values in the spectrum, we can calculate the average angle of the data points with respect to the vertical as well as determine if we want to rotate the image clockwise or counter-clockwise by that angle. If the slope of the diagonal is negative we will want to rotate the image clockwise, and vice versa with a positive slope.

Now that we have extracted an angle, we can actually try it out and see how our solution fared.

```python
rotatedBook1=scipy.misc.imrotate(Book1, -angle)
rotatedBook2=scipy.ndimage.interpolation.rotate(Book1, -angle)
plot_array(rotatedBook1)
plot_array(rotatedBook2)
```

The pages in the rotated photos are actually pretty close to being horizontal, off by perhaps a degree or two. This result is heavily influenced by precisely where we threshold the spectrum. 

It is also interesting to think of the image in terms of local direction of the text. There are portions of the curved surface of the page where the text is locally angled more downwards, and other sections where is is angled upwards. If we were to split the image up into smaller portions and calculate the necessary local rotation of the chunk, it is possible we could correct for the curvature of the page as well as more accurately correct for the orientation with the FT (though we would have to stitch the image back together intelligently). 

However, we would run into an unavoidable limit as we sliced the image into smaller and smaller grids--we would have less and less distinct diagonals in the spectrum to pick out as our grid resolution increased, a 2-d Gabor limit of sorts.

At this point in the image processing, there may already be Optical Character Recognition algorithms robust enough to parse this text, so long as all the local sections are close enough to horizontal. If there still needs to be futher processing to correct for the curvature, it will probably involve some intelligent interpolation and stitching sections of the image back together continuously, and would be difficult to do purely off of localized FTs.
