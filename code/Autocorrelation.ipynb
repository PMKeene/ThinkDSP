```python
def autocorr(wave):
    n = len(wave.ys)

    corrs = []
    lags = range(n//2)
    for lag in lags:
        y1 = wave.ys[lag:]
        y2 = wave.ys[:n-lag]
        corr = numpy.corrcoef(y1, y2)[0, 1]
        corrs.append(corr)

    return lags, corrs
```

```python
wave = thinkdsp.read_wave('28042__bcjordan__voicedownbew.wav')
wave.unbias()
wave.normalize()
wave.make_audio()
```

```python
segment1 = wave.segment(start=0.2, duration=0.01) #note: not the same section as correlated below
segment1.plot()
```

```python
n=512
ys = wave.ys[:n]
segment = thinkdsp.Wave(ys, wave.framerate) #Segment made of first 512 entries of the waveform
sample_lag, sample_corrs = autocorr(segment)
thinkplot.plot(sample_lag, sample_corrs)
```

```python
#Previously we visually examined the graph and invcestiggates corrs[80:90] for the peak
extracted=[sample_corrs[0]]
index=1
for n in sample_corrs[1:]:
    multiple=2
    elements=[n]
    while (index*multiple) < len(sample_corrs):
        elements.append(sample_corrs[index*multiple])
        multiple+=1
    extracted.append(sum(elements))
    index+=1
    
thinkplot.plot(sample_lag, extracted)
t=extracted.index(max(extracted))/segment.framerate
f=1/t
print(f)
```

```python
#Just wrapping all we did above into a function that returns the fundamental pitch of a segment
def Pitch(segment):
    lags, corrs = autocorr(segment)
    extracted=[corrs[0]]
    index=1
    for n in sample_corrs[1:]:
        multiple=2
        elements=[n]
        while (index*multiple) < len(sample_corrs):
            elements.append(sample_corrs[index*multiple])
            multiple+=1
        extracted.append(sum(elements))
        index+=1
    t=extracted.index(max(extracted))/segment.framerate
    return 1/t
    
Pitch(segment)
```
