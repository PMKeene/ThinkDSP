```python
import thinkdsp as dsp
import thinkplot
import math
import numpy
from __future__ import print_function, division
```

```python
PI2=math.pi*2
```

```python
class TriangleChirp(dsp.Chirp):

    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        cycles=phases/PI2

        frac, _ = numpy.modf(cycles)
        ys = numpy.abs(frac - 0.5)
        ys = dsp.normalize(dsp.unbias(ys), self.amp)
        return ys
```

```python
Mountains=TriangleChirp(start=300, end=400)
MountainRange=Mountains.make_wave()
```

```python
Mountain=MountainRange.segment(0,.01)
Mountain.plot()
```

```python
MountainRange.make_audio()
```

```python
Sauron=MountainRange.make_spectrum()
Sauron.plot()
```
